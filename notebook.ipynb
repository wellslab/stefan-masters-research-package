```python
import json 
import pandas as pd
from pprint import pprint

data = json.load(open('publication_data.json'))
df = pd.DataFrame.from_dict(data, orient='index')
df.head()
df.columns = ['journal']

```

```python

# Normalise the stem cell research journal names
def normalise_journal_name(record):
    journal = record['journal']
    if journal.lower() == "stem cell research":
        record['journal'] = "Stem Cell Research"
    return record

df_normalised = df.apply(normalise_journal_name, axis=1)   
    
```

```python
scr_pmids = []
non_scr_pmids = []

for pmid, record in df_normalised.iterrows():
    if record['journal'] == "Stem Cell Research":
        scr_pmids.append(pmid)
    else:
        non_scr_pmids.append(pmid)

```

```python
pprint(scr_pmids)
pprint(non_scr_pmids)
```
